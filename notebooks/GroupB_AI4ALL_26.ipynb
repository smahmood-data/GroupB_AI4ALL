```python
def num_outliers(df, column):
    try:
        mean = df[column].mean()
        std = df[column].std()
        outliers_df = df[df[column] > mean + 2 * std]
        return outliers_df.shape[0]
    except:
        return "N/A - invalid entries"
num_outliers(df, 'ndvi_ne')
```
